```python
import numpy as np
import pandas as pd
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

# Load Dataset

```python
df=pd.read_excel("Booking_details".xlsx")
```

# Checking Missing Values

```python
df.isnull().sum()/len(df)*100
```

# Drop Columns

```python
df.drop(columns=['Service Name', 'Service Type', 'Class Type', 'Instructor', 'Customer Email', 'Customer Phone','Subscription Type','Theme'],inplace=True)
```

```python
df
```

```python
df.isnull().sum()
```

# Filling Null Values

```python
df["Time Slot"].fillna(df["Time Slot"].mode()[0],inplace=True)
```

# Value counts

```python
df["Time Slot"].value_counts()
```

# Filling Null Values & Counts

```python
df["Duration (mins)"]=df["Duration (mins)"].fillna(df["Duration (mins)"].mean())
```

```python
df["Duration (mins)"].value_counts()
```

# Values round figure and Change datatype

```python
df['Duration (mins)'] = df['Duration (mins)'].round(0).astype(int)
```

# Rename column

```python
df.rename(columns={"Duration (mins)": "Duration(mins)"}, inplace=True)
```

```python
df
```

# Values round figure and change dtype

```python
df['Price'] = df['Price'].round(0).astype(int)
```

# Checking Value counts and filling null values

```python
df["Facility"].value_counts()
```

```python
df["Facility"].fillna("No Facility",inplace=True)
```

```python
df
```

```python
df.info()
```

# Change dtype and covert time 24hrs(am,pm)

```python
df["Time Slot"] = pd.to_datetime(df["Time Slot"], format="%H:%M:%S").dt.strftime("%I:%M %p")
```

```python
df
```

# first four letter from booking id and customer id

```python
df["Booking ID"] = df["Booking ID"].str[:4]
df["Customer ID"] = df["Customer ID"].str[:4]
```

```python
df
```

# Convert to csv file after cleaning

```python
df.to_csv("Booking_data.csv", index=False)
```
